### 1. Importación del Dataset con Pandas
- A continuación, se importará el dataset proporcionado utilizando Pandas


```python
import pandas as pd

df = pd.read_csv('calificaciones.csv')


```

### 2. Exploración Inicial de los Datos
- Empezamos a visualizar los datos del dataframe para saber con que datos vamos a trabajar

```python
print(df.head())
print(70*'*')
print(df.tail())
print(70*'*')
print(df.info())
print(70*'*')
print(df.isnull().sum())
```

### 3. Limpieza y Normalización de los Datos
- En esta parte del trabajo se realizaría la normalización y la corrección de los datos

```python
df.head()
```

```python
df.dropna(subset=['parcial1','parcial2','nota_final'], inplace=True)

df['fecha'] = pd.to_datetime(df['fecha'], errors='coerce') 
df['nombre'] = df['nombre'].str.capitalize()
df['apellido'] = df['apellido'].str.capitalize()

df.head()
```

### 4. Obtención de Estadísticas Descriptivas

- Se calculan estadísticas generales y ademas por grupos para así poder tener una mejor comprensión sobre los datos con los que estamos trabajando ahora


```python
print(df.describe())
print(70*'*')
print(df.groupby('aprobado').describe())
```

### 5. Visualización de los Datos con Matplotlib

- En este punto se realizará un gráfico de barras que permita ver la cantidad de estudiantes aprobados y reprobados 

```python
import matplotlib.pyplot as plt

df['aprobado'].value_counts().plot(kind='bar')
plt.title('Cantidad de Desaprobados y Aprobados')
plt.xlabel('Condición')
plt.ylabel('Cantidad')
plt.show()
```

### 6. Exportación a Archivo SQLite

- Acá guardaríamos el dataframe procesado en una base de datos SQLite y también verificaremos que se hayan guardado correctamente

```python
import sqlite3 


conexion = sqlite3.connect('calificaciones.sqlite')
df.to_sql('tabla_calificaciones', conexion, if_exists="replace", index=False)
df_cargado = pd.read_sql_query('SELECT * FROM tabla_calificaciones', conexion)
conexion.close()

df_cargado
```
